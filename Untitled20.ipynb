##Problem Statement 4: Hypothesis Testing
Objective: To use a t-test to determine if there is a statistically significant difference between two groups.
Dataset: The cleaned DataFrame from Problem 1.
Hypothetical Question: Is there a significant difference in the average price of Airbnb listings between Manhattan and Brooklyn?
Tasks:
State the Hypotheses:
Formulate the null hypothesis (H0), which assumes there is no difference in the average price between the two boroughs.
Formulate the alternative hypothesis (Ha), which assumes there is a difference.
Prepare the Data: Create two new Series or DataFrames: one containing the prices of listings in 'Manhattan' and another with the prices for listings in 'Brooklyn'.
Perform the Test:
Use the scipy.stats.ttest_ind() function to perform an independent two-sample t-test on the Manhattan and Brooklyn price data.
Interpret the Results:
Set a significance level (alpha) of =0.05.
Examine the p-value from your test result.
Based on the p-value, decide whether to reject or fail to reject the null hypothesis. State your conclusion in the context of the problem (e.g., "There is/is not a statistically significant difference in the average price...").


```python
df_cleaned["neighbourhood_group"].value_counts()
```

```python
## H0 = assumes there is no difference in the average price between the two boroughs.

## Ha = assumes there is a difference in the average price between the two boroughs.
```

```python
df1 = df_cleaned[df_cleaned["neighbourhood_group"] == "Manhattan"].price

df2 = df_cleaned[df_cleaned["neighbourhood_group"] == "Brooklyn"].price
```

```python
print("mean1" , df1.mean())
print("mean2" , df2.mean())

# We see the means are different
```

```python
print("n:" , df1.value_counts().sum())
print("n" ,  df2.value_counts().sum())

## the value of n is different
```

```python
print("variance1 :" , df1.var())
print("variance2 :" , df2.var())

#3 variance is different .
```

```python
from scipy import stats

stats.ttest_ind(df1, df2, equal_var = False)
```

```python
##p_value = 0.00 , aplha = 0.05 , we see alpha is more then p value, so we reject the null hypotheses.

## Hence their is a statistical difference between mean of values of mathatten and brooklyn.

```
